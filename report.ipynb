```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from kmeans import KMeansClusterClassifier
import random

df = pd.read_csv('C:\\Users\\Kemal\\Downloads\\Iris.csv')
df = df.drop("Id", axis = 1)
df = df.drop("Species", axis = 1)
df = df.drop("PetalWidthCm", axis = 1)
df_columns = df.columns.tolist()
list = df.values.tolist()

def train_test(df, test_size):
    index = df.index.tolist()
    test_idx = random.sample(population=index, k = test_size)
    test_df = df.loc[test_idx]
    train_df = df.drop(test_idx)
    return train_df, test_df

train_df, test_df = train_test(df, test_size = 20)
train_list = train_df.values.tolist()
train_list_with_columns = train_df.values.tolist()
train_list_with_columns.insert(0, df_columns)
test_list = test_df.values.tolist()

kmean = KMeansClusterClassifier(train_df.values)
points = kmean.fit(train_df.values)
clusters, elbow_errors = kmean.predict(points)
```

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d')
color=['red','blue','green','cyan','magenta','yellow','brown','black']
for i, c in enumerate(clusters):
            for p in c.points:
                ax.scatter(p.coords[0],p.coords[1],p.coords[2], c = color[i])
 # plot the point (2,3,4) on the figure

plt.show()
```
